## Step 2: 코랩 내장 Gemini에게 프롬프팅(질문) 하기

코랩 우측 상단의 **Gemini(✨)** 버튼을 누르고, 여러분의 아이디어를 바탕으로 코드를 수정해 달라고 요청하세요.

> **🤖 AI 질문(프롬프트) 예시:**
> "내가 만든 아래 파이썬 퀴즈 코드에 점수 계산 기능을 추가하고 싶어. 맞추면 10점, 틀리면 -5점이고, '종료'를 누르면 최종 점수를 출력하게 코드를 수정해 줘."
> `(여기에 Lab 2의 완성된 코드를 복사해서 붙여넣기)`

AI가 제시해 준 코드를 아래 빈 코드 셀에 복사해서 붙여넣고 **`Shift + Enter`**를 눌러 실제로 잘 동작하는지 테스트해 보세요!

```python
# [학생 작성란] Gemini가 수정해 준 '나만의 퀴즈 봇' 코드를 여기에 붙여넣고 실행해 보세요.
import pandas as pd
import time
import random

# Đọc dữ liệu từ file csv của bạn
try:
    df = pd.read_csv('my_words.csv', encoding='utf-8-sig')
    words = df.values.tolist()

    score = 0  # Biến lưu điểm số
    total_questions = 5  # Số câu hỏi mỗi lượt chơi

    print("--- BẮT ĐẦU QUIZ BOT: PHIÊN BẢN TÍNH ĐIỂM ---")
    print("Luật chơi: Đúng +10đ | Sai hoặc Quá 15s: -5đ")
    print("-" * 45)

    for i in range(total_questions):
        quiz = random.choice(words)
        print(f"\nCâu {i+1}: Từ '{quiz[0]}' có nghĩa là gì?")

        start_time = time.time()
        user_input = input("Trả lời: ")
        end_time = time.time()

        elapsed = end_time - start_time

        if user_input == quiz[1] and elapsed <= 15:
            score += 10
            print(f"✅ Tuyệt vời! (+10đ). Bạn mất {elapsed:.2f} giây.")
        elif elapsed > 15:
            score -= 5
            print(f"⏰ Quá 15 giây rồi! (-5đ). Đáp án đúng là: {quiz[1]}")
        else:
            score -= 5
            print(f"❌ Sai mất rồi! (-5đ). Đáp án đúng là: {quiz[1]}")

        print(f"Điểm hiện tại của bạn: {score} điểm")

    print("-" * 45)
    print(f"TRÒ CHƠI KẾT THÚC! TỔNG ĐIỂM CUỐI CÙNG: {score} ĐIỂM")

except Exception as e:
    print("Lỗi: Hãy đảm bảo bạn đã tải file 'my_words.csv' lên ở Step 1.")
```
